```python
# классификация логрег
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_curve, auc
import matplotlib.pyplot as plt

df_human = pd.read_csv("dimensions_human.csv")
df_bot1 = pd.read_csv("dimensions_bot1.csv")
df_bot2 = pd.read_csv("dimensions_bot2.csv")
df_bot3 = pd.read_csv("dimensions_bot3.csv")
df_bot4 = pd.read_csv("dimensions_bot4.csv")

# выбор ботов для train и test
df_bot_train = pd.concat([df_bot1, df_bot2], ignore_index=True)
df_bot_test = pd.concat([df_bot3, df_bot4], ignore_index=True)

# делим human на train и test
df_human_train, df_human_test = train_test_split(
    df_human,
    test_size=0.3,  # 30% в тест
    random_state=42,
    stratify=df_human["label"]
)

# сбор выборки
df_train = pd.concat([df_human_train, df_bot_train], ignore_index=True)
df_test = pd.concat([df_human_test, df_bot_test], ignore_index=True)

X_train = df_train[["D1"]]
y_train = df_train["label"]

X_test = df_test[["D1"]]
y_test = df_test["label"]

model = LogisticRegression(class_weight='balanced', random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

df_human = pd.read_csv("dimensions_human.csv")
df_bot1 = pd.read_csv("dimensions_bot1.csv")
df_bot2 = pd.read_csv("dimensions_bot2.csv")
df_bot3 = pd.read_csv("dimensions_bot3.csv")
df_bot4 = pd.read_csv("dimensions_bot4.csv")

df_bot_train = pd.concat([df_bot1, df_bot2], ignore_index=True)
df_bot_test = pd.concat([df_bot3, df_bot4], ignore_index=True)


df_human_train, df_human_test = train_test_split(
    df_human, test_size=0.3, random_state=42, stratify=df_human["label"]
)


df_train = pd.concat([df_human_train, df_bot_train], ignore_index=True)
df_test = pd.concat([df_human_test, df_bot_test], ignore_index=True)


X_train = df_train[["D1"]] 
y_train = df_train["label"]
X_test = df_test[["D1"]]
y_test = df_test["label"]


clf = DecisionTreeClassifier(random_state=42)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")


print("\nClassification Report:")
print(classification_report(y_test, y_pred))


print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```

Тест Уилкоксона

```python
import pandas as pd

df_human = pd.read_csv("dimensions_human.csv")
human_data = df_human['D1']  

bot_files = [
    "dimensions_bot1.csv",
    "dimensions_bot2.csv",
    "dimensions_bot3.csv",
    "dimensions_bot4.csv"
]


bot_data_list = []
for file in bot_files:
    df = pd.read_csv(file)
    bot_data_list.append(df['D1'])

df_bots_combined = pd.concat(bot_data_list, ignore_index=True)

```

```python
from scipy.stats import mannwhitneyu

stat, p_value = mannwhitneyu(human_data, df_bots_combined, alternative='two-sided')

print("p-value:", p_value)
if p_value < 0.05:
    print("Значимое различие между ботами и людьми.")
else:
    print("Нет значимого различия.")
```
